```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
Apple = yf.download('AAPL', start='2010-01-01', end='2023-05-01')
Meta = yf.download('META', start='2010-01-01', end='2023-05-01')
Tesla = yf.download('TSLA', start='2010-01-01', end='2023-05-01')
Pfizer = yf.download('PFE', start='2010-01-01', end='2023-05-01')
Sea_Limited = yf.download('SE', start='2010-01-01', end='2023-05-01')
FTSE_100 = yf.download('^FTSE', start='2010-01-01', end='2023-05-01')
Dow_Jones_Industrial_Average = yf.download('^DJI', start='2010-01-01', end='2023-05-01')
NASDAQ_Composite = yf.download('^IXIC', start='2010-01-01', end='2023-05-01')
Russell_2000 = yf.download('RUT', start='2010-01-01', end='2023-05-01')
Nikkei_225 = yf.download('^N225', start='2010-01-01', end='2023-05-01')
```

# For Apple Inc.

```python
Apple['Daily Returns'] = Apple['Close'].pct_change()
returns = Apple['Daily Returns'] + 1
cumulative_returns = returns.cumprod()
Apple['Cumulative Returns'] = cumulative_returns
Apple['Rolling Max'] = Apple['Cumulative Returns'].rolling(window=len(Apple), min_periods=1).max()
Apple['Drawdown'] = Apple['Cumulative Returns'] - Apple['Rolling Max']
max_drawdown = Apple['Drawdown'].min()
average_return = Apple['Daily Returns'].mean()*252
standard_deviation = Apple['Daily Returns'].std()*np.sqrt(252)
volatility = np.sqrt(252) * Apple['Daily Returns'].std()
risk_free_rate = 0.035
sharpe_ratio = (average_return-risk_free_rate)/standard_deviation
downside_returns = Apple['Daily Returns'][Apple['Daily Returns'] < 0]
downside_deviation = downside_returns.std()
average_return = Apple['Daily Returns'].mean()
sortino_ratio = (average_return - risk_free_rate) / downside_deviation
print('Volatility:', volatility)
print('Max Drawdown:', max_drawdown)
print('Sharpe Ratio:', sharpe_ratio)
print('Sortino Ratio:', sortino_ratio)
Apple
```
