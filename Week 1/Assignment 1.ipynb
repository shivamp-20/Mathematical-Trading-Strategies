```python
returns1=data1.pct_change()
returns2=data2.pct_change()
```

```python
start_date = '2010-01-01'
end_date = returns1.index[-1].strftime('%Y-%m-%d')
```

```python
returns1 = returns1.loc[start_date:end_date]
returns2 = returns2.loc[start_date:end_date]
```

```python
cummulative_return1=(1+returns1).cumprod()-1
cummulative_return2=(1+returns2).cumprod()-1
```

```python
cumm1=cummulative_return1*100
cumm2=cummulative_return2*100
```

    MAX DRAWDOWN

```python
for column in data1.columns[0:]:
    Max = data1[column].cummax()
    Daily_Drawdown = data1[column]/Max - 1.0
    Max_Daily_Drawdown = Daily_Drawdown.cummin()
    k=Max_Daily_Drawdown.mean()*100
    print(column + " : ", '{:.6f}%'.format(k))
```

```python
for column in data2.columns[0:]:
    Max = data2[column].cummax()
    Daily_Drawdown = data2[column]/Max - 1.0
    Max_Daily_Drawdown = Daily_Drawdown.cummin()
    k=Max_Daily_Drawdown.mean()*100
    print(column + " : ", '{:.6f}%'.format(k))
```
