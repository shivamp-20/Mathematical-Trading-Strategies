```python
import pandas as pd
import yfinance as yf
import numpy as mp
```

```python
def alldone(t):
    t_data = yf.download(t, start = '2010-01-01', end = '2023-05-01')
    t_closeprices = t_data['Close']
    t_dailyreturns = t_closeprices.pct_change()
    t_volatility = mp.sqrt(252) * t_dailyreturns.std()
    t_cumreturns = (1+t_dailyreturns).cumprod() - 1
    t_cumreturnmean = t_cumreturns.mean()
    t_rollmax = t_closeprices.cummax()
    t_dailydrawdown = t_closeprices/t_rollmax - 1
    t_maxdd = t_dailydrawdown.cummin()
    risk_free = 0.0
    t_sharpe = mp.sqrt(252) * ((t_dailyreturns.mean() - risk_free)/t_dailyreturns.std())
    t_downside = t_dailyreturns[t_dailyreturns < 0]
    t_sortino = mp.sqrt(252) * ((t_dailyreturns.mean() - risk_free)/t_downside.std())
    print("Daily Returns : ")
    print(t_dailyreturns.tail())
    print("\nVolatality : ", t_volatility)
    print("\nCumulative Returns : ")
    print(t_cumreturns.tail())
    print("\nMean Cum Return : ", t_cumreturnmean)
    print("\nMax Draw Down : ")
    print(t_maxdd.tail())
    print("\nSharpe Ratio : ")
    print(t_sharpe)
    print("\nSortino Ratio : ")
    print(t_sortino)
```

```python
print("Equities - \n\n\n")
print("Google : \n")
alldone('GOOGL')
print("\n\nApple : \n")
alldone('AAPL')
print("\n\nAmazon : \n")
alldone('AMZN')
print("\n\nMicrosoft : \n")
alldone('MSFT')
print("\n\nMeta : \n")
alldone('META')
```
