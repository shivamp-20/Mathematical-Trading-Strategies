```python
#FOR INDICES

import yfinance as yf
import numpy as np
import statistics as st

tickers=["^gspc","^dji","^ixic","^nya","^xax"]  

def cum_return(arr1,arr2):    
    a=len(arr1)
    cr=(arr1[len(arr1)-1]-arr2[0])/arr2[0]
    return cr

def maxd(trough,peak):
    max=(trough-peak)/peak   
    return max

cumulative_return=[]            
volatility=[]        
sharpe_ratio=[]       
sortino_ratio=[]     
max_drawdown=[]       

for x in range(5):   
    neg=[]           

    df =yf.download(tickers[x],  start = "2010-01-01" , end = "2023-05-16")    
    adj_close,close,open=df['Adj Close'],df['Close'],df['Open']                
    cumulative_return.append(cum_return(close,open))                   
    
    adjClose=np.array(adj_close)
    opn=np.array(open)                
    diff=np.subtract(adjClose,opn)             
    difference=list(diff) 
    for y in range(len(difference)):    
        if difference[y]<0:           
            neg.append(difference[y])
            
    neg_stdev=st.stdev(neg)       
    stdev=st.stdev(difference)
    volatility.append(stdev*(np.sqrt(252)))   
    
    mean=st.mean(difference)
    sharpe_ratio.append(((mean*100)-7.07)/stdev)    
    
    sortino_ratio.append(((mean*100)-7.07)/neg_stdev)
    
    possible=[]
    for i in range(len(close)):
        for j in range(len(close)):
            if i<j and close[i]>=close[j]:
                possible.append((maxd(close[j],close[i]))*100)    #max drawdown in %
    mx=123456789            
    for x in range(len(possible)):
        if possible[x]<=mx:
            mx=possible[x]
    max_drawdown.append(mx)    

print(cumulative_return)
print(volatility)
print(sharpe_ratio)
print(sortino_ratio)
print(max_drawdown)
```
