```python
import yfinance as yf
import pandas as pd
import numpy as np
import math
```

```python
equities=["TSLA","ABT","ADBE","AAPL","MSFT"]
indices=["^GSPC","^BSESN","^DJI","^RUT","^FTSE"]
data=yf.download(tickers=equities[0], start="2010-01-01", end="2023-05-01", interval="1d")
df=pd.DataFrame(data)
# print(df)
Daily_return=((df["Close"]-df["Close"].shift(1))/df["Close"].shift(1))*100*252
print(Daily_return/252)
print(Daily_return)
risk_free_return = 4
excess_return = Daily_return - risk_free_return
print(excess_return)
```

```python
for t in range(5):
    print(equities[t])
    data=yf.download(tickers=equities[t], start="2010-01-01", end="2023-05-01", interval="1d")
    df=pd.DataFrame(data)
    #CUMMULATIVE RETURNS PERCENTAGE
    #CUMMULATIVE RETURNS PERCENTAGE
    ts=df[['Adj Close']]
    cum=((ts.iloc[-1]-ts.iloc[0])/ts.iloc[0])
    print(cum)
    
    ##MAXIMUM DRAWDOWN RATIO
    peak = df.expanding(min_periods=1)["Close"].max()
    max_drawdown = (data["Close"]-peak)/peak
    max_draw_down=max_drawdown.min()
    print(max_draw_down)
    
    ##DAILY RETURNS
    Daily_return=((df["Close"]-df["Close"].shift(1))/df["Close"].shift(1))*100*252
    
#     print(Daily_return)
#     sharpe_ratio
    #Treasury rate of USA on 1 May 2023 was 4.86%.
    risk_free_return = 4
    excess_return = Daily_return - risk_free_return
    sharpe_ratio = (np.mean(excess_return)/np.std(excess_return))*(math.sqrt(252))
    print(sharpe_ratio)
    
    #Volatility
    volatility=np.std(Daily_return)*(math.sqrt(252))
    print(volatility)
    
    #downside deviations
    down_dev_crit=Daily_return[Daily_return<0]
#     sq=np.square(down_dev_crit)
#     down_dev=(math.sqrt(np.sum(sq)/len(Daily_return)))
    down_dev=np.std(down_dev_crit)
    print(down_dev)

    #sortino
    sortino=(np.mean(excess_return)/down_dev)*(252**(1/2))
    print(sortino)

    t=t+1

```
