```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
def sharperatio(data, riskrate):

    returns = (data['Close']- data['Open']) / data['Open']
    excessreturns = returns - riskrate
    average_excessreturns = np.mean(excessreturns)
    standard_deviation = np.std(returns)
    sharpe_ratio = average_excessreturns / standard_deviation
    
    return sharpe_ratio
# sharperatio(data, 0.0)
```

```python
def maxdrawdown(data):
    prices = data['Close']
    peak = prices[0]
    max_drawdown = 0.0

    for price in prices:
        if price > peak:
            peak = price
        drawdown = (peak - price) / peak
        if drawdown > max_drawdown:
            max_drawdown = drawdown

    return max_drawdown


# maxdrawdown(data)
```

```python
def returns(data):
  currentvalue = data['Close'][-1]
  originalprice = data['Close'][0]

  return (currentvalue - originalprice)/ originalprice

# returns(data)
```

```python
def sortinoratio(data, riskrate, targetreturn):
   
    returns = np.diff(data['Close']) / data['Close'][:-1]
    excessreturns = returns - riskrate
    target_returns = excessreturns[excessreturns < targetreturn]
    targetdeviation = np.std(target_returns)
    sortino_ratio = (np.mean(target_returns) - targetreturn) / targetdeviation
    
    return sortino_ratio
# sortinoratio(data, 0,0)
```

```python
def volatility(data):
    
    returns = np.log(data['Close'] / np.roll(data['Close'], 1))
    returns = returns[1:]
    volatility = np.std(returns)
    
    return volatility
# volatility(data)
```

```python
def dailyreturns(data):
   
   DailyReturn = (data['Close']- data['Open']) *100/ data['Open']
   return sum(DailyReturn)

#dailyreturns(data)
```

```python
dat = ['Cumlitative Returns(%)', 'Volatility', 'Sharpe Ratio', 'Sortiono Ratio', 'Max Drawdown(%)']
equity_table = pd.DataFrame(index = dat)
equitylist = ['RELIANCE.NS', 'TATASTEEL.NS', 'ADANIGREEN.NS', 'AMBUJACEM.NS', 'ASIANPAINT.NS','TCS.NS','NESTLEIND.NS']
riskrate = 0.07
targetreturn = 0
for equity in equitylist:
  ticker = yf.Ticker(equity)
  data = ticker.history(interval='1d', start='2010-01-01', end='2023-05-01')
  equity_table[equity] = [returns(data) * 100,volatility(data), sharperatio(data, riskrate), sortinoratio(data, riskrate, targetreturn), maxdrawdown(data) * 100]
equity_table
```
