```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
import plotly.express as px
import plotly.graph_objs as go
```

```python
tickers = ['VFIAX','JNK','DIA','SCHA','NASDX']
```

```python
df = pd.DataFrame()
for ticker in tickers:
    df_tmp=yf.download(ticker, start='2010-1-1', end='2023-5-10')
    df_tmp['Ticker'] = ticker
    df = pd.concat([df, df_tmp])
df
```

```python
df.reset_index(inplace=True)
```

```python
df.groupby('Ticker')['Date'].agg(['min', 'max', 'count'])
```

```python
df.isnull().sum()
```

```python
df_mod = df[['Date', 'Ticker', 'Adj Close']]
# make columns names prettier
df_mod.columns = ['date', 'ticker', 'price']
df_mod
```

```python
df1 = df_mod.pivot_table(index=['date'], columns='ticker', values=['price'])
# flatten columns multi-index, `date` will become the dataframe index
df1.columns = [col[1] for col in df1.columns.values]
df1_reuse=df1
df1
```

Compute cumulative returns from prices

```python
cum_return = (df1.iloc[-1] - df1.iloc[0]) / df1.iloc[0]
cum_return * 100
```

Compute cumulative returns from simple daily returns

```python
# compute daily returns using pandas pct_change()
df_daily_returns = df1.pct_change()
# skip first row with NA 
df_daily_returns = df_daily_returns[1:]
df_daily_returns
```

```python
# Calculate the cumulative daily returns
df_cum_daily_returns = (1 + df_daily_returns).cumprod() - 1
df_cum_daily_returns = df_cum_daily_returns.reset_index()
df_cum_daily_returns
```

```python
cum_return_entire_period = df_cum_daily_returns.iloc[:, 1:].tail(1)
cum_return_entire_period * 100
```

Visualize the returns

```python
# reset the index, moving `date` as column
df_daily_returns = df_daily_returns.reset_index()
# use `melt`
df1 = df_daily_returns.melt(id_vars=['date'], var_name='ticker', value_name='daily_return')
# add one more column, showing the daily_return as percent
df1['daily_return_pct'] = df1['daily_return'] * 100
df1
```

```python
fig = px.line(df1, x='date',
              y='daily_return_pct', color='ticker',
              title='Performance - Daily Simple Returns',
              labels={'daily_return_pct':'daily returns (%)'})
fig.show()
```

```python
df2 = df_cum_daily_returns.melt(id_vars=['date'], var_name='ticker', value_name='cum_return')
df2['cum_return_pct'] = df2['cum_return'] * 100
df2
```

```python
fig = px.line(df2, x='date',
              y='cum_return_pct', color='ticker',
              title='Performance - Daily Cumulative Returns',
              labels={'cum_return_pct':'daily cumulative returns (%)', })
fig.show()
```

```python
df1_mod = df1[['date', 'ticker', 'daily_return']]
df1_mod
```

```python
df3 = df1_mod.pivot_table(index=['date'], columns='ticker', values=['daily_return'])
# flatten columns multi-index, `date` will become the dataframe index
df3.columns = [col[1] for col in df3.columns.values]
df3
```

the daily, monthly, and annually *Volatility* of a stock

```python
import math
for column in df3.columns[0:]:
    daily_volatility = df3[column].std()
    print(column)
    print('Daily volatility: ', '{:.2f}%'.format(daily_volatility*100))
    monthly_volatility = math.sqrt(21) * daily_volatility *100
    print('Monthly volatility: ', '{:.2f}%'.format(monthly_volatility))
    annual_volatility = math.sqrt(252) * daily_volatility *100
    print('Annual volatility: ', '{:.2f}%'.format(annual_volatility))
```

Sharpe Ratio

```python
for column in df3.columns[0:]:
    daily_volatility = df3[column].std()
    # print(column)
    Sharpe_Ratio =(df3[column].mean()*252) / (df3[column].std()*np.sqrt(252))
    
    print(column + " : ", '{:.6f}'.format(Sharpe_Ratio))
```

Sortiono Ratio

```python
for column in df3.columns[0:]:
    downside_returns = df3.loc[df3[column] < 0]
    expected_return = df3[column].mean()
    down_stdev = downside_returns[column].std()
    rfr=0
    sortino_ratio = (expected_return - rfr)/down_stdev
    sortino_ratio=sortino_ratio*np.sqrt(252)
    print(column + " : ", '{:.6f}'.format(sortino_ratio))


    
```

Max Drawdown

```python
for column in df1_reuse.columns[0:]:
    Roll_Max = df1_reuse[column].cummax()
    Daily_Drawdown = df1_reuse[column]/Roll_Max - 1.0
    Max_Daily_Drawdown = Daily_Drawdown.cummin()
    k=Max_Daily_Drawdown.mean()*100
    print(column + " : ", '{:.6f}%'.format(k))
    
```
