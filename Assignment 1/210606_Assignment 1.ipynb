Choose 5 international indices and 5 equities from yfinance, load the daily data since “2010-01-01”-"2023-05-01", and compute the Daily Returns, Cumulative Returns, Max Drawdowns, the Sharpe Ratio, and the Sortino Ratio if the stocks were bought, held for the day and sold when the market closed.

```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
Indices = ['^DJI','^NYA','^IXIC','^XAX','^GSPC']
Equities = ['goog','dpz','tsla', 'amzn','msft']
```

```python
indices_data = yf.download(Indices,start='2010-01-01',end='2023-05-01')
equities_data = yf.download(Equities,start='2010-01-01',end='2023-05-01')
```

```python
indices_data
```

```python
daily_return_indices = (indices_data['Close']-indices_data['Open'])/indices_data['Open']
daily_return_equities = (equities_data['Close']-equities_data['Open'])/equities_data['Open']
```

```python
daily_return_indices
```

```python
cumulative_return_indices = (indices_data.iloc[-1]['Close'] - indices_data.iloc[0]['Open'])/indices_data.iloc[0]['Open']
cumulative_return_indices
```

```python
cumulative_return_equities = (equities_data.iloc[-1]['Close'] - equities_data.iloc[0]['Open'])/equities_data.iloc[0]['Open']
cumulative_return_equities
```

calculation of maximum draw down

```python
# for indices
max_indices_data = indices_data['Adj Close'].rolling(window=len(indices_data['Adj Close']),min_periods=1).max()
dd_indices = indices_data['Adj Close']/max_indices_data - 1
MDD_indices = dd_indices.rolling(window=len(indices_data['Adj Close']),min_periods=1).min()
MDD_indices.min()*100
```

```python
# for equities
max_equities_data = equities_data['Adj Close'].rolling(window=len(equities_data['Adj Close']),min_periods=1).max()
dd_equities = equities_data['Adj Close']/max_equities_data - 1
MDD_equities = dd_equities.rolling(window=len(equities_data['Adj Close']),min_periods=1).min()
MDD_equities.min()*100
```

sharpe ratio = (average_return - risk_free_return)
               ------------------------------------
                       standard deviation

```python
standard_deviation_indices = daily_return_indices.std()
standard_deviation_equities = daily_return_equities.std()

print(standard_deviation_indices)
print(standard_deviation_equities)
```

```python
time_period = indices_data.shape[0]
```

```python
import math
volatility_indices = standard_deviation_indices*math.sqrt(time_period)
volatility_equities = standard_deviation_equities*math.sqrt(time_period)
print(volatility_indices)
print(volatility_equities)
```

```python
rfr = 0.02
excess_indices = daily_return_indices - rfr
excess_equities = daily_return_equities - rfr

sharpe_indices = excess_indices.mean()/excess_indices.std()
sharpe_equities = excess_equities.mean()/excess_equities.std()

print(sharpe_indices)
print(sharpe_equities)
```

```python
# sortino ratio
neg_return_indices = daily_return_indices[daily_return_indices < 0]
neg_return_equities = daily_return_equities[daily_return_indices < 0]

sortino_indices = excess_indices.mean()/neg_return_indices.std()
sortino_equities = excess_equities.mean()/neg_return_equities.std()

print(sortino_indices)
print(sortino_equities)
```
