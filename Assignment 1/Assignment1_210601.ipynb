```python
import yfinance as yf
import pandas as pd
import numpy as np

start_date = "2010-01-01"
end_date = "2023-05-01"

equities = ["AAPL", "AMZN", "NFLX", "INTC", "AMD"]
# equity_data = yf.download(equities, start=start_date, end=end_date)
# # equity_adj_close=equity_data["Adj Close"]
# # equity_data1 = yf.download("AAPL", start=start_date, end=end_date)

indices=["^GSPC", "^N100", "IMOEX.ME", "^BSESN", "^IXIC" ]
# index_data=yf.download(indices, start=start_date, end=end_date )
# m=equity_data.count()[0] #3353

# def std_dev(x):
#   x['Returns']=x['Adj Close'].pct_change()
#   std_dev=x['Returns'].std()
#   return std_dev

#cumulative returns
def cum_ret(x):
  #here, x is Adj Close feature of the stock
  current_price=x.iloc[-1]
  original_price=x.iloc[0]
  cr=(current_price-original_price)/original_price
  return cr

#max drawdown
def mdd(x):
  #here x is equivalent to aapl=yf.download("AAPL", start=start_date, end=end_date)
  peak=x['High'].max()
  trough=x['Low'].min()
  percentage_difference = ((trough - peak) / peak) * 100
  return percentage_difference

#sharpe ratio
def sharpe_ratio(x):
  #here x is equivalent to aapl=yf.download("AAPL", start=start_date, end=end_date)
  x["DailyReturn"] = x["Close"].pct_change()
  risk_free_rate = 0.0525  # Risk-free rate, for stocks based in USA 5.25% is assumed here
  index_return = x["DailyReturn"].mean() * 252  # Annualized return, assuming 252 trading days in a year
  index_volatility = x["DailyReturn"].std() * np.sqrt(252)  # Annualized volatility
  sharpe_ratioo = (index_return - risk_free_rate) / index_volatility
  return sharpe_ratioo

#sortino ratio
def sortino_ratio(x):
  #here x is equivalent to aapl=yf.download("AAPL", start=start_date, end=end_date)
  x["DailyReturn"] = x["Close"].pct_change()
  risk_free_rate = 0.0525
  downside_returns = x.loc[x["DailyReturn"] < 0, "DailyReturn"]
  downside_volatility = downside_returns.std() * np.sqrt(252)  # Annualized downside volatility
  index_return = x["DailyReturn"].mean() * 252
  sortino_ratioo = (index_return - risk_free_rate) / downside_volatility
  return sortino_ratioo

for equity in equities:
  equity_data=yf.download(equity, start=start_date, end=end_date)
  # aapl=equity_data['Adj Close']['AAPL']
  
  CumulativeReturn=cum_ret(equity_data['Adj Close'])
  print(f"Cumulative Return of {equity} is: {CumulativeReturn}")
  
  maxDrawDown=mdd(equity_data)
  print(f"Maximum drawdown of {equity} is: {maxDrawDown}")
  
  SharpeRatio=sharpe_ratio(equity_data)
  print(f"Sharpe Ratio of {equity} is: {SharpeRatio}")
  
  SortinoRatio=sortino_ratio(equity_data)
  print(f"Sortino Ratio of {equity} is: {SortinoRatio}")

  volatility=equity_data["DailyReturn"].std() * np.sqrt(252) *100
  print(f"Annual Volatility of {equity} is: {volatility}")

for index in indices:
  index_data=yf.download(index, start=start_date, end=end_date)
  # aapl=index_data['Adj Close']['^N100']
  
  CumulativeReturn=cum_ret(index_data['Adj Close'])
  print(f"Cumulative Return of {index} is: {CumulativeReturn}")
  
  maxDrawDown=mdd(index_data)
  print(f"Maximum drawdown of {index} is: {maxDrawDown}")
  
  SharpeRatio=sharpe_ratio(index_data)
  print(f"Sharpe Ratio of {index} is: {SharpeRatio}")
  
  SortinoRatio=sortino_ratio(index_data)
  print(f"Sortino Ratio of {index} is: {SortinoRatio}")

  volatility=index_data["DailyReturn"].std() * np.sqrt(252) *100
  print(f"Annual Volatility of {index} is: {volatility}")
```
