```python
import yfinance as yf
import pandas as pd
import numpy as np

international_indices_equities = ['^GSPC', '^IXIC', '^DJI', '^N225', '^HSI','AMZN', 'NSRGF', 'GOOGL', 'MSFT', 'TSLA']


for index in international_indices_equities :
    index_info = yf.download(index, start="2010-01-01", end="2023-05-01")
    index_info.reset_index(inplace=True)
    
    index_info['Daily_Returns'] = ((index_info['Close'] - index_info['Open']) / index_info['Open']) * 100
    
    index_info['Cummulative_Returns'] = ((index_info['Daily_Returns'][0]) / 100) + 1
    
    for i in range(1, len(index_info)):
        index_info.loc[i, 'Cummulative_Returns'] = (index_info.loc[i-1, 'Cummulative_Returns']) * (((index_info.loc[i, 'Daily_Returns']) / 100) + 1)
    
    index_info['sample'] = index_info['Cummulative_Returns']
    index_info['Cummulative_Returns'] = (index_info['Cummulative_Returns'] - 1) * 100
    
    volatility_in_percent = index_info['Daily_Returns'].std()
    
    risk_free_rate = 0
    
    sharpe_ratio = (index_info['Daily_Returns'].mean() - risk_free_rate) / volatility_in_percent
    
    negative_values = index_info['Daily_Returns'][index_info['Daily_Returns'] < 0]
    downward_deviation = negative_values.std()
    
    sortino_ratio = (index_info['Daily_Returns'].mean() - risk_free_rate) / downward_deviation
    
    index_info['peak'] = index_info['sample'][0]
    
    for i in range(1, len(index_info)):
        if index_info['sample'][i] > index_info['peak'][i-1]:
            index_info.loc[i, 'peak'] = index_info['sample'][i]
        else:
            index_info.loc[i, 'peak'] = index_info['peak'][i-1]
    
    index_info['drawdown'] = ((index_info['sample']) - (index_info['peak'])) * 100
    
    MDD = index_info['drawdown'].min()
    
    cummulative_returns = index_info['Cummulative_Returns'].iloc[-1]
    
 
    
    
    # Print the results for each index
    print(f"Index: {index}")
    print("Cummulative Returns:",cummulative_returns)
    print("Volatility:", volatility_in_percent)
    print("Sharpe Ratio:", sharpe_ratio)
    print("Sortino Ratio:", sortino_ratio)
    print("Maximum Drawdown:", MDD)
  
    print("---------------------------------------------------------")
    
```
