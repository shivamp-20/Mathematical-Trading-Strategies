# Importing Libraries

```python
import yfinance as yf
import numpy
import pandas as pd
```

# Data of Indices and Equity
Downloading Data from Yfinance library

```python
Tickers=['AAPL','MSFT','AMZN','GE','GOOG']
ticker_2=['^GSPC','^DJI','^IXIC','^N225','^VIX']
Equity=yf.download(Tickers,start='2010-01-01')
Index=yf.download(ticker_2,start='2010-01-01')
print(Equity)
print(Index)
```


# Maximum Drawdown

MDD = $\frac{Trough Value-Peak Value}{Peak Value}$


```python
MDD_equity=[(min(Equity['Low'][i])-max(Equity['High'][i]))/(max(Equity['High'][i]))*100 for i in Tickers]
MDD_index=[(min(Index['Low'][i])-max(Index['High'][i]))/(max(Index['High'][i]))*100 for i in ticker_2]
MDD_equity_Dict= {key: round(value,2) for key,value in zip(Tickers,MDD_equity)}
MDD_index_dict = {key: round(value,2) for key, value in zip(ticker_2,MDD_index)}
print(MDD_index_dict)
MDD_equity_Dict
```

# Cumulative Returns

Cumulative Return = $\left(\frac{{\text{Current Price-Original Price}}}{{\text{Original Price}}}\right)\times 100$


```python
Cumulative_ret_equity=[round((((Equity['Adj Close'][i][-2]/Equity['Adj Close'][i][0])-1))*100,2) for i in Tickers ]
Cumulative_ret_index=[((((Index['Adj Close'][i][-2])/(Index['Adj Close'][i][0])))-1)*100 for i in ticker_2 ]
Cumulative_ret_dict_equity={key: round(value,2) for key, value in zip(Tickers,Cumulative_ret_equity)}
Cumulative_ret_dict_index={key: round(value,2) for key, value in zip(ticker_2,Cumulative_ret_index)}
print(Cumulative_ret_dict_equity)
Cumulative_ret_dict_index
```

#
Calculating *daily Returns* for annulising purpose

```python
daily_ret_equity=[[],[],[],[],[]]
daily_ret_index=[[],[],[],[],[]]
for i in range(1,len(Equity['Adj Close'])):
    k=0
    for j in Tickers:
        daily_ret_equity[k].append(round((((Equity['Adj Close'][j][i]-Equity['Adj Close'][j][i-1])/(Equity['Adj Close'][j][i-1]))*100),2))
        k=k+1
for i in range(1,len(Index['Adj Close'])):
    k=0
    for j in ticker_2:
        daily_ret_index[k].append(round((((Index['Adj Close'][j][i]-Index['Adj Close'][j][i-1])/(Index['Adj Close'][j][i-1]))*100),2))
        k=k+1
daily_ret_dict_equity={key: value for key, value in zip(Tickers, daily_ret_equity)}
daily_ret_dict_index={key: value for key, value in zip(ticker_2, daily_ret_index)}
daily_ret_dict_equity
```

# 
Calculating Daily median return and annualising it 
#

```python
daily_median_ret_equity=[numpy.nanmedian(daily_ret_equity[i]) for i in range(0,5)]
Annual_return_equity=[(pow((1+(daily_median_ret_equity[i]/100)),365)-1)*100 for i in range(0,5)]
daily_median_ret_index=[numpy.nanmedian(daily_ret_index[i]) for i in range(0,5)]
Annual_return_index=[(pow((1+(daily_median_ret_index[i]/100)),365)-1)*100 for i in range(0,5)]
Annual_return_dict_equity={key: round(value,2) for key, value in zip(Tickers,Annual_return_equity)}
Annual_return_dict_index={key: round(value,2) for key, value in zip(ticker_2,Annual_return_index)}
print(Annual_return_dict_index)
print(Annual_return_dict_equity)
```

**VOLATILITY**
#
Its the annual standard deviation 
#
σ<sub>a</sub> = σ<sub>daily</sub> &radic;T


```python
annual_std_dev_equity=[round(numpy.nanstd(daily_ret_equity[i])*pow(len(daily_ret_equity[i])/365,0.5),2) for i in range(0,5)]
Volatility_equity={key: value for key, value in zip(Tickers,annual_std_dev_equity)}
annual_std_dev_index=[round(numpy.nanstd(daily_ret_index[i])*pow(len(daily_ret_index[i])/365,0.5),2) for i in range(0,5)]
Volatility_index={key: value for key, value in zip(ticker_2,annual_std_dev_index)}
print(Volatility_index)
Volatility_equity
annual_std_dev_equity
```

## Sharpe Ratio
for Equity = $\frac{Annual Return  of Equity - Risk free Rate }{\sigma_a}$

for Index =  $\frac{Annual Return  of Index - Risk free Rate }{\sigma_a}$



```python
Risk_free_rate_US = 4.25
sharpe_ratio_equity = [(Annual_return_equity[i]-Risk_free_rate_US)/annual_std_dev_equity[i] for i in range(0,5)]
sharpe_ratio_equity_dict = {key: round(value,2) for key, value in zip(Tickers,sharpe_ratio_equity)}
sharpe_ratio_index = [(Annual_return_index[i]-Risk_free_rate_US)/annual_std_dev_index[i] for i in range(0,5)]
sharpe_ratio_index_dict = {key: round(value,2) for key, value in zip(ticker_2,sharpe_ratio_index)}
print(sharpe_ratio_equity_dict)
sharpe_ratio_index_dict
```

# Downside Deviation #
Downside deviation is calculated for calculating Sortino Ratio. \
Downside deviation is the standard deviation for all the negative daily returns. \
The median downside deviation is then annualised by the formula \
σ<sub>annual</sub> = σ<sub>daily</sub> &radic;T

```python
for j in range(0,5):
    for i in daily_ret_index[j]:
        if i>0:
            daily_ret_index[j].remove(i)
for j in range(0,5):
    for i in daily_ret_equity[j]:
        if i>0:
            daily_ret_equity[j].remove(i)
annual_downside_std_dev_equity=[round(numpy.nanstd(daily_ret_equity[i])*pow(len(daily_ret_equity[i])/365,0.5),2) for i in range(0,5)]
annual_downside_std_dev_index=[round(numpy.nanstd(daily_ret_index[i])*pow(len(daily_ret_index[i])/365,0.5),2) for i in range(0,5)]
```

# Sortino Ratio
 for Equity = $\frac{Annual Return  of Equity - Annual Return of Index }{\sigma_d}$ \
 \
 for Index = $\frac{Annual Return  of Index - Risk free Rate}{\sigma_d}$

```python
sortino_ratio_equity=[]
sortino_ratio_index=[]
i=0
for value in Annual_return_dict_equity.values():
    sortino_ratio_equity.append(round((value-Annual_return_dict_index['^GSPC'])/annual_downside_std_dev_equity[i],2))
    i=i+1
i=0
for key,value in Annual_return_dict_index.items():
    sortino_ratio_index.append(round((value-Risk_free_rate_US)/annual_downside_std_dev_index[i],2))
    i=i+1
sortino_ratio_index
```
