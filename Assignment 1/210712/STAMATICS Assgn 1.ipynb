```python
import yfinance as yf
import math
import pandas as pd
```

# EQUITIES

```python
data = yf.download("AAPL GOOGL NVDA OLED AMZN", start="2010-01-01", end="2023-05-01", group_by="ticker")
data
```

```python
length=len(data)
length
```

```python
stock_list=["AAPL", "GOOGL", 'NVDA', 'OLED', 'AMZN']
```

## Cumulative Returns

```python
CR_dict={}

for i in stock_list:
    end=data[i]["Close"][length-1]
    start=data[i]["Open"][0]
    CR_dict[i]= (end-start)/start
    print("CR for", i, ":", CR_dict[i])
```

## Daily Returns / Volatility

```python
DR_dict={}
for i in range(0,5):
    DR_dict[i]= data[stock_list[i]]["Close"]-data[stock_list[i]]["Open"]
    DR_dict[i]=DR_dict[i].to_frame()
    DR_dict[i]=DR_dict[i].rename(columns={0:'dr'})
```

```python
DR_dict
```

```python
DR_mean_dict={}

for i in range(5):
    DR_mean_dict[i]=DR_dict[i]["dr"].mean()

DR_mean_dict
```

```python
for i in range(5):
    DR_dict[i]["diff"]=DR_dict[i]["dr"]-DR_mean_dict[i]
    DR_dict[i]["sqdiff"]=DR_dict[i]["diff"]*DR_dict[i]["diff"]
DR_dict
```

```python
var_dict={}
annual_vol_dict={}

for i in range(5):
    var_dict[i]= DR_dict[i]["sqdiff"].sum()/length
    annual_vol_dict[i]=math.sqrt(var_dict[i]*252)
print("Variances:\n", var_dict)
print("Annualised Volatilities:\n", annual_vol_dict)
```

## Sharpe Ratio

assuming risk-free returns is 5.25% [USA] for all the indices

```python
sharpe_ratio_dict={}

for i in range(5):
    sharpe_ratio_dict[i]= (DR_mean_dict[i]*252-0.0525)/annual_vol_dict[i]

print("Sharpe Ratios:\n", sharpe_ratio_dict)
```

## Sortino Ratio

```python
neg_DR_dict={}

for i in range(5):
    neg_DR_dict[i] = DR_dict[i][DR_dict[i]['dr'] < 0]

neg_DR_dict
```

```python
neg_DR_mean_dict={}

for i in range(5):
    neg_DR_mean_dict[i]=neg_DR_dict[i]["dr"].mean()

neg_DR_mean_dict
```

```python
for i in range(5):
    neg_DR_dict[i]['negdiff']=neg_DR_dict[i]['dr']-neg_DR_mean_dict[i]
    neg_DR_dict[i]['negsqdiff']=neg_DR_dict[i]['negdiff']*neg_DR_dict[i]['negdiff']

neg_DR_dict
```

```python
neg_var_dict={}

for i in range(5):
    neg_var_dict[i]=neg_DR_dict[i]['negsqdiff'].mean()

neg_var_dict
```

```python
annual_neg_stddev_dict={}

for i in range(5):
    annual_neg_stddev_dict[i]=math.sqrt(252*neg_var_dict[i])

annual_neg_stddev_dict
```

```python
sortino_ratio_dict={}

for i in range(5):
    sortino_ratio_dict[i]= (DR_mean_dict[i]*252-0.0525)/annual_neg_stddev_dict[i]

sortino_ratio_dict
```

## Maximum Draw-Down

```python
new_dict={}

for i in range(5):
    df1=data[stock_list[i]]['Open']
    df2=data[stock_list[i]]['Close']
    new_dict[i]=pd.concat([df1,df2])
    new_dict[i].sort_index()

new_dict
```

```python
for i in range(5):
    new_dict[i]=new_dict[i].to_frame()
for i in range(5):
    new_dict[i]['1']=new_dict[i].shift(-1)

new_dict
```

```python
for i in range(5):
    new_dict[i]['increase_bool']=new_dict[i][0].le(new_dict[i]['1'])

new_dict
```

```python
for i in range(5):
    new_dict[i]['shift']=new_dict[i]['increase_bool'].shift(-1)

new_dict
```

```python
for i in range(5):
    new_dict[i] = new_dict[i].astype({"shift":'bool'})
```

```python
for i in range(5):
    new_dict[i]['shift_invert']=(~new_dict[i]['shift'])

new_dict
```

```python
for i in range(5):
    new_dict[i]['peak_bool']= new_dict[i]['increase_bool']&new_dict[i]['shift_invert']

new_dict
```

```python
dd_list_dict={}

for i in range(5):
    crest=0
    dd_list_dict[i]=[]
    dd=0
    for index, row in new_dict[i].iterrows():
        if row['peak_bool']:
            crest=row['1']
            dd_list_dict[i]+=[dd]
            dd=0
        else :
            trough=row[0]
            dd=(crest-trough)/trough

dd_list_dict
```

```python
max_dd_dict={}

for i in range(5):
    max_dd_dict[i]=max(dd_list_dict[i])

max_dd_dict
```
