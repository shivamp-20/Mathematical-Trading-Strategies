```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
tickers =  ['MA', 'JPM', 'AAPL', 'AMZN', 'F','^IXIC','^GSPC','CAC','DAX','BABA']
df = yf.download(tickers, start='2010-01-01', end='2023-01-01')
```

```python
df
```

```python
df1=yf.download(tickers, start='2010-01-01', end='2023-01-01')['Adj Close']
df1
```

```python
cumulative_return = (df1.iloc[-1] - df1.iloc[0])/df1.iloc[0]
cumulative_return
```

```python
cumulative_return*100
```

```python
def calculate_max_drawdown(prices):
    cum_returns = prices / prices.iloc[0]
    previous_peaks = cum_returns.cummax()
    drawdowns = cum_returns - previous_peaks
    max_drawdown = drawdowns.min()
    return max_drawdown

max_drawdowns = df1.apply(calculate_max_drawdown, axis=0)
max_drawdowns
```

```python
N = 255 #total trading days in a year
rf=0.01 #risk free rate
def sharpe_ratio(return_series, N, rf):
    mean = return_series.mean()* N -rf
    sigma = return_series.std()* np.sqrt(N)
    return mean/sigma

sharpe = df1.apply(sharpe_ratio, args=(N,rf,),axis=0)
sharpe
```

```python
def calculate_sortino_ratio(prices, risk_free_rate=0.0):
    returns = prices.pct_change().dropna()
    downside_returns = np.where(returns < risk_free_rate, returns, 0)
    downside_deviation = np.std(downside_returns)
    expected_return = np.mean(returns)
    sortino_ratio = (expected_return - risk_free_rate) / downside_deviation
    return sortino_ratio

sortino_ratios = df1.apply(calculate_sortino_ratio, risk_free_rate=0.02)
sortino_ratios
```

```python
def vola(return_series, N, rf):
  return return_series.std() * np.sqrt(N)

vola = df1.apply(vola, args=(N,rf,),axis=0)
vola
```

```python
ratios = ["cumulative_return","max_drawdowns", "sharpe", "sortino_ratios", "vola"]

ratio_dict = {
    ratios[0]: cumulative_return,
    ratios[1]: max_drawdowns,
    ratios[2]: sharpe,
    ratios[3]: sortino_ratios,
    ratios[4]: vola,
}
df_final = pd.DataFrame(ratio_dict)
df_final.index =['MA', 'JPM', 'AAPL', 'AMZN', 'F','^IXIC','^GSPC','CAC','DAX','BABA']
df_final
```
