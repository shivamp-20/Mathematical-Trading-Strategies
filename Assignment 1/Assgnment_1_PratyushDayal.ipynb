```python
df2_filter["price_t-1"] = df2_filter["price_t"].shift(1)
df2_filter["daily_returns"] = (df2_filter["price_t"]/ df2_filter["price_t-1"]) - 1

df3_filter["price_t-1"] = df3_filter["price_t"].shift(1)
df3_filter["daily_returns"] = (df3_filter["price_t"]/ df3_filter["price_t-1"]) - 1

df4_filter["price_t-1"] = df4_filter["price_t"].shift(1)
df4_filter["daily_returns"] = (df4_filter["price_t"]/ df4_filter["price_t-1"]) - 1

df5_filter["price_t-1"] = df5_filter["price_t"].shift(1)
df5_filter["daily_returns"] = (df5_filter["price_t"]/ df5_filter["price_t-1"]) - 1
```

```python
df1_filter=df1_filter.dropna()
df2_filter=df2_filter.dropna()
df3_filter=df3_filter.dropna()
df4_filter=df4_filter.dropna()
df5_filter=df5_filter.dropna()
```

```python
def sharpe_ratio(data, risk_free_rate=0.03):
    # Calculate Average Daily Return
    mean_daily_return = data.mean()*252-0.03
    # Calculate Standard Deviation
    sigma = data.std()*math.sqrt(252)
    return mean_daily_return/sigma


#df1_filter["daily_returns"].std()
#sum(df1_filter["daily_returns"])/len(df1_filter["daily_returns"])
```

```python
print("HDFCBANK sharpe Ratio:",sharpe_ratio(df1_filter["daily_returns"]))
print("Tesla sharpe Ratio:",sharpe_ratio(df2_filter["daily_returns"]))
print("Tyo sharpe Ratio:",sharpe_ratio(df3_filter["daily_returns"]))
print("BoeingCo sharpe Ratio:",sharpe_ratio(df4_filter["daily_returns"]))
print("IOC sharpe Ratio:",sharpe_ratio(df5_filter["daily_returns"]))
```

```python
def sortino_ratio(data, risk_free_rate=0.03):
    # Calculate Average Daily Return
    mean_daily_return = data.mean()*252-0.03
    # Calculate Standard Deviation
    std_neg = data[data<0].std()*math.sqrt(252)
    return mean_daily_return/std_neg
sortino_ratio(df1_filter["daily_returns"])
```

```python
print("HDFCBANK sortino Ratio:",sortino_ratio(df1_filter["daily_returns"]))
print("Tesla sortino Ratio:",sortino_ratio(df2_filter["daily_returns"]))
print("Tyo sortino Ratio:",sortino_ratio(df3_filter["daily_returns"]))
print("BoeingCo sortino Ratio:",sortino_ratio(df4_filter["daily_returns"]))
print("IOC sortino Ratio:",sortino_ratio(df5_filter["daily_returns"]))
```

```python
#now international indices

tickers=["NDAQ","DAX","^BSESN","^GSPC","^FTSE"]
df=pd.DataFrame()
for ticker in tickers:
    df_tick=yf.download(ticker,start="2010-01-01",end="2023-05-01")
    df_tick["Ticker"]=ticker
    df=pd.concat([df,df_tick])
df.dropna()   
```

```python
df1 = df.pivot_table(index=['Date'], columns='Ticker', values=['Adj Close'])
# flatten columns multi-index, `date` will become the dataframe index
df1.columns = [col[1] for col in df1.columns.values]
df1
```

```python
df1_daily_returns = df1.pct_change()
# skip first row with NA 
df1_daily_returns = df1_daily_returns[1:]
df1_daily_returns
```

```python
df1_cum_daily_returns = (1 + df1_daily_returns).cumprod() - 1
df1_cum_daily_returns = df1_cum_daily_returns.reset_index()
df1_cum_daily_returns
```

```python
cum_return_entire_period = df1_cum_daily_returns.iloc[:, 1:].tail(1)
cum_return_entire_period * 100
```
