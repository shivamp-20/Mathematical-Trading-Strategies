```python
import yfinance as yf
import plotly.graph_objects as go

class Stock:

    def __init__(self, symbol):
        self.symbol = symbol  # Replace with your desired stock symbol
        self.data = yf.download(self.symbol, start="2010-01-01", end="2023-05-01")


    def plot_line(self):
        self.data.Close.plot(title=self.symbol);

    def plot_candlesticks(self):
        # Create a new column to track color based on price change
        self.data['Color'] = 'red'  # Default color is red
        self.data.loc[self.data['Close'] > self.data['Open'], 'Color'] = 'green'  # Change color to green when close > open

        # Create candlestick trace
        candlestick = go.Candlestick(
            x=self.data.index,
            open=self.data['Open'],
            high=self.data['High'],
            low=self.data['Low'],
            close=self.data['Close'],
            increasing=dict(line=dict(color='green')),
            decreasing=dict(line=dict(color='red'))
        )
         # Create layout
        layout = go.Layout(
            title=f'{self.symbol} Stock Price',
            yaxis=dict(title='Price')
        )

        # Create figure
        fig = go.Figure(data=[candlestick], layout=layout)

        # Display the figure
        fig.show()
```

```python
ptc = Stock("PTC")
ptc.plot_candlesticks()
#asending triangle
```

```python
axs = Stock("AXS")
axs.plot_candlesticks()
#HEAD AND SHOULDER
```

```python
abeo = Stock("ABEO")
abeo.plot_candlesticks()
#BEAR FLAG
```

```python
acco = Stock("ACCO")
acco.plot_candlesticks()
#rounding top
```

```python
adnt = Stock("ABEV")
adnt.plot_candlesticks()
#cup and handle
```
