```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# List of stock symbols
stock_symbols = [ 'GOOGL', 'JPM', 'MSFT', 'AAPL', 'AMZN', 'NFLX', 'TSLA', 'V', 'WMT', 'JNJ']

df_stocks = pd.DataFrame()

# historical data for each stock
for symbol in stock_symbols:
    stock = yf.Ticker(symbol)
    stock_data = stock.history(period='10y')
    stock_data['Symbol'] = symbol
    df_stocks = pd.concat([df_stocks, stock_data])  
```

Descending Triangle

```python

# check if a descending triangle pattern is present
def is_descending_triangle(data):
    highs = data['High']
    lows = data['Low']
    close = data['Close']

    if len(highs) < 4:
        return False

    if highs[0] >= highs[1] and lows[1] >= lows[2] and highs[3] >= highs[2]:
        if lows[0] <= lows[1] and close[0] > close[1]:
            return True

    return False

marked_regions = {}

# Iterate over the stock symbols to identify descending triangles
for symbol in stock_symbols:
    stock_data = df_stocks[df_stocks['Symbol'] == symbol]
    marked_dates = []
    for i in range(len(stock_data) - 3):
        data_subset = stock_data.iloc[i:i+4]
        if is_descending_triangle(data_subset):
            marked_dates.append(data_subset.index[0])
    marked_regions[symbol] = marked_dates

# Plot the stock graphs with marked regions
for symbol in stock_symbols:
    stock_data = df_stocks[df_stocks['Symbol'] == symbol]
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label=symbol)
    for date in marked_regions[symbol]:
        plt.scatter(date, stock_data.loc[date, 'Close'], marker='v', color='r')
    plt.title(symbol + ' Stock')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    # marked dates 
    if len(marked_regions[symbol]) > 0:
        print(f"Marked dates for {symbol}:")
        for date in marked_regions[symbol]:
            print(date)
        print("--------------------")
```

Bull Flag

```python
# check if a bull flag pattern is present
def is_bull_flag(data):
    highs = data['High']
    lows = data['Low']
    close = data['Close']

    if len(highs) < 4:
        return False

    if highs[0] <= highs[1] <= highs[2] and lows[0] <= lows[1] <= lows[2]:
        if close[0] < close[1] < close[2]:
            return True

    return False

marked_regions = {}

# Iterate over the stock symbols to identify bull flag patterns
for symbol in stock_symbols:
    stock_data = df_stocks[df_stocks['Symbol'] == symbol]
    marked_dates = []
    for i in range(len(stock_data) - 3):
        data_subset = stock_data.iloc[i:i+4]
        if is_bull_flag(data_subset):
            marked_dates.append(data_subset.index[0])
    marked_regions[symbol] = marked_dates

# Plot the stock graphs with marked regions
for symbol in stock_symbols:
    stock_data = df_stocks[df_stocks['Symbol'] == symbol]
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label=symbol)
    for date in marked_regions[symbol]:
        plt.scatter(date, stock_data.loc[date, 'Close'], marker='^', color='g')
    plt.title(symbol + ' Stock')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    if len(marked_regions[symbol]) > 0:
        print(f"Marked dates for {symbol}:")
        for date in marked_regions[symbol]:
            print(date)
        print("--------------------")
```

Inverse Head and Shoulder
