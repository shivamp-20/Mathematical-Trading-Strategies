```python
import numpy as np
import pandas as pd
import yfinance as yf
from math import nan, sqrt
import matplotlib.pyplot as plt
import matplotlib
```

```python
stocks=['AMZN', 'GOOG', 'META', 'NVDA', 'AAPL']
sdt=yf.download(" ".join(stocks), start='2010-01-01', end='2023-06-20')
dt=pd.DataFrame(sdt)
dt
```

```python
dt.columns
```

```python
plt.figure(figsize=(15,10))
plt.plot(dt.index, dt[('Adj Close', 'AMZN')], '-b', label='AMZN')
plt.plot(dt.index[2600:3100], [185]*500, '-r', label='Rounding top')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot(dt.index, dt[('Adj Close', 'GOOG')], '-b', label='GOOG')
plt.plot(dt.index[3100:3200], [105]*100, '-r', label='Bear flag')
plt.plot(dt.index[3100:3200], [123]*100, '-r',)
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot(dt.index, dt[('Adj Close', 'META')], '-b', label='META')
plt.plot(dt.index[2650:3050], [250]*400, '-r', label='Head and Shoulders')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot(dt.index, dt[('Adj Close', 'NVDA')], '-b', label='NVDA')
plt.plot(dt.index[3050:3360], [110]*310, '-r', label='Cup and Handle')
plt.plot(dt.index[3050:3360], [280]*310, '-r')
plt.legend()
```

```python
plt.figure(figsize=(15,10))
plt.plot(dt.index[:], dt[('Adj Close', 'AAPL')][:], '-b', label='AAPL')
plt.plot(dt.index[2650:2900], [105]*250, '-r', label='Ascending Triangle')
plt.plot(dt.index[2650:2900], [140]*250, '-r')

plt.legend()
```
