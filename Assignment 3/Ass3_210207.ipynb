```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the list of stock tickers
stock_tickers = ["AAPL", "MSFT", "AMZN", "GOOGL", "FB", "TSLA", "JPM", "JNJ", "PG", "V"]

# Iterate over each stock ticker
for ticker in stock_tickers:
    # Download historical stock data
    stock_data = yf.download(ticker, start="2011-01-01", end="2021-01-01")
    
    # Extract the necessary data for plotting
    dates = stock_data.index
    close_prices = stock_data["Close"]

    # Plot the price chart
    plt.figure(figsize=(12, 6))
    plt.plot(dates, close_prices)
    plt.title("Price Chart - " + ticker)
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.grid(True)
    plt.show()
```
