```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Apple stock data for the past year
symbol = "AAPL"
data = yf.download(symbol, start="2020-01-01", end="2020-04-01")

# Extract necessary data for plotting
high_prices = data["High"]
low_prices = data["Low"]
close_prices = data["Close"]

# Plotting the descending triangle
plt.figure(figsize=(10, 6))
plt.plot(close_prices, color="black", label="Close Price")
plt.plot(high_prices, color="green", linestyle="--", label="High Price")
plt.plot(low_prices, color="red", linestyle="--", label="Low Price")

# Draw the triangle lines
start_point = high_prices.idxmax()
end_point = low_prices.idxmin()
plt.plot([start_point, end_point], [high_prices[start_point], high_prices[end_point]], color="blue", linewidth=2)
plt.plot([end_point, end_point], [high_prices[end_point], low_prices[end_point]], color="blue", linewidth=2)

# Set chart title and labels
plt.title(f"Descending Triangle - {symbol}", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price", fontsize=12)

# Add legend and grid
plt.legend(loc="upper right")
plt.grid(True)

# Display the chart
plt.show()
```
