## Keltner Channel

```python
def Keltner_Channel(index, atr_period, emr_period, multiplier):
    true_range=np.maximum(data[index]["High"]-data[index]["Low"], abs(data[index]["High"]-data[index]["Close"].shift()), abs(data[index]["Close"].shift()-data[index]["Low"]))
    true_range.fillna(0)

    average_true_range=true_range.rolling(atr_period).mean()

    middle_line=data[index]["Close"].ewm(span=emr_period).mean()
    middle_line.rename("Middle")
    upper_line=middle_line+multiplier*average_true_range
    upper_line.rename("Upper")
    lower_line=middle_line-multiplier*average_true_range
    lower_line.rename("Lower")

    KC=data[index]
    KC["Upper"]=upper_line
    KC["Middle"]=middle_line
    KC["Lower"]=lower_line

    return(KC)
```

```python
Keltner_Channel("^IXIC",9,20,2)
```

```python
Keltner_Channel("^NSEI", 9, 20, 2)
```

## Bollinger Bands

```python
def Bollinger_Bands(index, periods, multiplier):
    BB=data[index]
    BB['Moving Average']=BB["Close"].rolling(window=periods).mean()
    BB['Standard Deviation']=BB["Close"].rolling(window=periods).std()
    BB["Upper"]=BB['Moving Average']+multiplier*BB['Standard Deviation']
    BB["Lower"]=BB['Moving Average']-multiplier*BB['Standard Deviation']
    
    return(BB)
```

```python
Bollinger_Bands("^IXIC", 20, 2)
```

```python
Bollinger_Bands("^NSEI", 20, 2)
```

## MACD Indicator

```python
def MACD(index, fast_period, slow_period, signal_period):
    MACD=data[index]
    ema_fast=MACD["Close"].ewm(span=fast_period).mean()
    ema_slow=MACD["Close"].ewm(span=slow_period).mean()
    MACD["MACD Line"]= ema_fast-ema_slow
    MACD["Signal Line"]=MACD["MACD Line"].ewm(span=signal_period).mean()
    MACD["MACD Histogram"]=MACD["MACD Line"]-MACD["Signal Line"]
    
    return(MACD)
```

```python
MACD("^IXIC", 12, 26, 9)
```

```python
MACD("^NSEI", 12, 26, 9)
```

# TASK 4 - Parameter Optimisation

We optimise based on NASDAQ because it is the lagging indicator.

## KC

```python
ema_periods=range(5, 51, 5)
atr_periods=range(5, 51, 5)
multipliers=list(range(1,7))
multipliers=[i * 0.5 for i in multipliers]
```

```python
CR_nasdaq_KC={}
for i in ema_periods:
    for j in atr_periods:
        for k in multipliers:
            kc=Keltner_Channel("^IXIC", i, j, k)
            kc['Signal']=np.where(kc['Close']>kc['Upper'], -1, np.where(kc['Close']<kc["Lower"], 1, 0))
            kc['Lag Signal']=kc['Signal'].shift()
            
            kc["Returns"]=kc['Lag Signal']*kc["Close"].pct_change()
            kc["Cumulative Returns"]=(kc["Returns"]+1).cumprod()
            
            CR=kc["Cumulative Returns"][-1]
            
            CR_nasdaq_KC.update({(i,j,k):CR})
CR_nasdaq_KC
```

```python
best_CR_KC= max(CR_nasdaq_KC.values())
best_CR_KC
```

```python
opt_para_KC=list(filter(lambda x: CR_nasdaq_KC[x] == best_CR_KC, CR_nasdaq_KC))
opt_para_KC
```
