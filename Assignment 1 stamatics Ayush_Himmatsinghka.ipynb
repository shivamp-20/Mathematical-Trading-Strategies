```python
import pandas as pd

import pandas_datareader as webreader
!pip install datetime
import datetime 
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
start = datetime.datetime(2010,1,1)
end = datetime.datetime(2023,5,1)
```

```python
stocks_indices = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'GOLDBEES.NS','^NSEI', '^DJI', '^IXIC', '^GSPC', '^RUT'] #selecting stocks and indices
```

```python
df = pd.DataFrame()
for stock in stocks_indices:
#     df_tmp = webreader.DataReader(stock,'yahoo',start,end)
    df_tmp = yf.download(stock,start,end)
    df_tmp['Stock'] = stock
    df=pd.concat([df,df_tmp])
df=df.reset_index()
df
```

```python
df.groupby("Stock")['Date'].agg(['min','max','count'])
```

```python
df1=df[['Date','Stock','Adj Close']]
df1.columns= ['Date', 'Stock','Price']
df1.head(5)
```

```python
df1_pivot = df1.pivot_table(index=['Date'],columns='Stock',values=['Price'])
df1_pivot.columns = [col[1] for col in df1_pivot.columns.values]
df1_pivot.head(5)
```

```python
cum_return =(df1_pivot.iloc[-1] - df1_pivot.iloc[0])/df1_pivot.iloc[0]
cum_return*100
```

```python
daily_returns = df1_pivot.pct_change()
daily_returns = daily_returns[1:]
daily_returns.head(5)
```

```python
window = 252
Roll_Max = df1['Price'].rolling(window, min_periods=1).max()
Daily_Drawdown = df1['Price']/Roll_Max - 1.0
Max_Daily_Drawdown = Daily_Drawdown.rolling(window, min_periods=1).min()
Max_Daily_Drawdown
```

```python
Sharpe_Ratio= daily_returns.mean()/daily_returns.std()
Sharpe_Ratio
```

```python
# weights = [1/10, 1/10, 1/10, 1/10, 1/10, 1/10, 1/10, 1/10, 1/10, 1/10]
# returns = df.pct_change(1).dropna()
# exp_rets = returns.mean()
# mean = sum(exp_rets * weights)
# var = np.dot(np.dot(weights, returns.cov()), weights)
# sr = mean / np.sqrt(var)
# sr
df2=df1_pivot.pct_change().dropna()
def sortino_ratio(series, N,rf):
    mean = series.mean() * N -rf
    std_neg = series[series<0].std()*np.sqrt(N)
    return mean/std_neg
N = 255 #255 trading days in a year
rf =0.01 #1% risk free rate
Sortino_Ratio = df2.apply(sortino_ratio, args=(N,rf,), axis=0 )
Sortino_Ratio
```
